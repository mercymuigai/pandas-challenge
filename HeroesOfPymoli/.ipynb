```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python

# File to Load 
HeroesOfPymoli_file = "Resources/purchase_data.csv"



# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(HeroesOfPymoli_file)


purchase_data.head()

```

```python
# Player Count

total_players = len(purchase_data["SN"].value_counts())

player_count = pd.DataFrame({"Total Players":[total_players]})

player_count
```

```python
#Purchasing Analysis (Total)

number_of_unique_items = purchase_data["Item ID"].nunique()

Total_Revenue = purchase_data["Price"].sum()

Average_Price = purchase_data["Price"].mean()

Number_of_Purchases = purchase_data["Item Name"].count()

# Creating a Dataframe
purchasing_analysis = pd.DataFrame ({"Number of Unique Items": [number_of_unique_items],"Average Price": [Average_Price],
                                      "Number of Purchase": [Number_of_Purchases],
                                      "Total Revenue":[Total_Revenue]})

purchasing_analysis["Average Price"]= purchasing_analysis["Average Price"].map("${:.2f}".format)

purchasing_analysis
```

```python

# # Gender Demographics

Group_gender = purchase_data.groupby("Gender")["SN"].nunique()


gender_percentage = Group_gender / total_players * 100


# Data Frame for Demographics

Gender_demographics = pd.DataFrame({"Total Count": Group_gender,"Percentage of Players": gender_percentage})

#Formating
Gender_demographics["Percentage of Players"]= Gender_demographics["Percentage of Players"].map("${:.2f}".format)


Gender_demographics
```

```python
# Purchasing Analysis (Gender)
 
    # Purchase Count
    #Average Purchase Price
    #Total Purchase Price
    #Avg Total Purchase Per Person

Purchase_Count = purchase_data["Gender"].value_counts()

grouped_gender_APP = purchase_data.groupby("Gender")["Price"].mean()

grouped_gender_TPV = purchase_data.groupby("Gender")["Price"].sum()

Avg_Total_Purchase_Per_Person = grouped_gender_TPV/Group_gender

# DataFrame for Purchasing Analysis

Purchasing_Analysis = pd.DataFrame({"Purchase Count":Purchase_Count,
                                    "Average Purchase Price":grouped_gender_APP,
                                    "Total Purchase Price":grouped_gender_TPV,
                                    "Avg Total Purchase Per Person":Avg_Total_Purchase_Per_Person})

#top Index Format
Gender_demographics.index.name = "Gender"

#formating



Purchasing_Analysis["Average Purchase Price"] = Purchasing_Analysis["Average Purchase Price"].map("${:.2f}".format)

Purchasing_Analysis["Total Purchase Price"] = Purchasing_Analysis["Total Purchase Price"].map("${:.2f}".format)

Purchasing_Analysis["Avg Total Purchase Per Person"] = Purchasing_Analysis["Avg Total Purchase Per Person"].map("${:.2f}".format)


Purchasing_Analysis
```

```python
# # Purchasing Analysis (Age)

# #Binning Purchase_Data by age
# #Purchase Count
# #Avg. Purchase Price
# #Total Purchase Value
# #Avg.Purchase total per Person by age group

#Creating bins for ages

bins = [0,9, 14, 19, 24, 29, 34, 39, 100]

# Labels for the bins
age_ranges = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=age_ranges)


# Groupby Age Group


age_groupby = purchase_data.groupby("Age Group")["SN"].nunique()

age_groupby

purchase_count_age = age_groupby["Purchase ID"].count()

Avg_Purchase_Price = age_groupby["Price"].mean()

Total_Purchase_Value = (purchase_count_age * Avg_Purchase_Price)

#Creating Data Frame

Purchasing_Analysis = pd.DataFrame({"Purchase Count":purchase_count_age, 
                                    "Average Purchase Price":Avg_Purchase_Price,
                                    "Total Purchase Value": Total_Purchase_Value})
#Formating 


Purchasing_Analysis["Average Purchase Price"] = Purchasing_Analysis["Average Purchase Price"].map("${:.2f}".format)

Purchasing_Analysis["Total Purchase Value"] = Purchasing_Analysis["Total Purchase Value"].map("${:.2f}".format)

Purchasing_Analysis
```
