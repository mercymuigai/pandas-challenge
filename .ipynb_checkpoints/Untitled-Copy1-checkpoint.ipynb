```python
# Dependencies and Setup
import pandas as pd
```

```python

# File to Load 
HeroesOfPymoli_file= "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_pd = pd.read_csv(HeroesOfPymoli_file)

purchase_data_pd.head()
```

```python
# Player Count

total_players = len(purchase_data_pd["SN"].value_counts())

player_count = pd.DataFrame({"Total Players":[total_players]})

player_count
```

```python
#Purchasing Analysis (Total)

number_of_unique_items = purchase_data_pd["Item ID"].nunique()

Total_Revenue = purchase_data_pd["Price"].sum()

Average_Price = purchase_data_pd["Price"].mean()

Number_of_Purchases = purchase_data_pd["Item Name"].count()

# Creating a Dataframe
purchasing_analysis = pd.DataFrame ({"Number of Unique Items": [number_of_unique_items],"Average Price": [Average_Price],
                                      "Number of Purchase": [Number_of_Purchases],
                                      "Total Revenue":[Total_Revenue]})
purchasing_analysis
```
